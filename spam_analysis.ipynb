```python
#Stop words removal, this task takes place during the data filtering, the purpose is to remove all the words that may appear in spam email


data = pd.read_csv('mail_data.csv')

message = ''.join(data['Message'].astype(str).tolist())

sw = set(STOPWORDS)
sw.update(["Click below", "virus", "awarded", "Congratulation", "Save up to", "won", "claim", "u"])

wc = WordCloud(background_color="grey", stopwords=sw, max_words=10000)

#display the words in lower case
wordcloud = wc.generate(message.lower())

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#Splitting data into variables
X = data['Message']
y = data['Category']
```

```python
#Splitting data into testing and training data 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.20)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#Converting text into numbers 
cv = CountVectorizer() 
x_train = cv.fit_transform(X_train)
x_test = cv.transform(X_test)
```

# Radial basis function kernel


```python
#Train the model using Support Vector Machine with radial basis function kernel(RBF)
model_rbf = SVC(kernel="rbf")
model_rbf.fit(x_train, y_train)

y_pred_rbf = model_rbf.predict(x_test)

accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
precision_rbf = precision_score(y_test, y_pred_rbf, average='weighted')
recall_rbf = recall_score(y_test, y_pred_rbf, average='weighted')
f1_rbf = f1_score(y_test, y_pred_rbf, average='weighted')
```

```python
print("Accuracy for RBF Kernel:", accuracy_rbf)
print("Precision for RBF Kernel:", precision_rbf)
print("Recall for RBF Kernel:", recall_rbf)
print("F1 Score for RBF Kernel:", recall_rbf)
```

# RBF Kernel with Gamma and C

```python
#Train the model using Support Vector Machine with radial function kernel(RBF)
model_rbf_type_2 = SVC(kernel= "rbf",gamma=15, C=7 ,random_state = 0)
model_rbf_type_2.fit(x_train, y_train)

y_pred_rbf_gamma_and_c = model_rbf_type_2.predict(x_test)


accuracy_rbf_gamma_and_c = accuracy_score(y_test, y_pred_rbf_gamma_and_c)
precision_rbf_gamma_and_c = precision_score(y_test, y_pred_rbf_gamma_and_c, average='weighted')
recall_rbf_gamma_and_ = recall_score(y_test, y_pred_rbf_gamma_and_c, average='weighted')
f1_rbf_gamma_and_ = f1_score(y_test, y_pred_rbf_gamma_and_c, average='weighted')
```

```python
print("Accuracy for Kernel with Gamma and C :", accuracy_rbf_gamma_and_c)
print("Precision for Kernel with Gamma and C:", precision_rbf)
print("Recall for Kernel with Gamma and C :", recall_rbf)
print("F1 Score for Kernel with Gamma and C :", recall_rbf)
```

# Linear Kernel

```python
#Train and fit the model using Support Vector Machine with linear kernel
model_linear_kernel = SVC(kernel= "linear",random_state = 0)
model_linear_kernel.fit(x_train, y_train)  

y_pred_linear = model_linear_kernel.predict(x_test)

accuracy_linear = accuracy_score(y_test, y_pred_linear)
precision_linear = precision_score(y_test, y_pred_linear, average='weighted')
recall_linear = recall_score(y_test, y_pred_linear, average='weighted')
f1_linear = f1_score(y_test, y_pred_linear, average='weighted')
```

```python
print("Accuracy for Linear Kernel: ", accuracy_linear)
print("Precision for RBF Kernel: ", precision_linear)
print("Recall for RBF Kernel: ", recall_linear)
print("F1 Score for RBF Kernel: ", f1_linear)
```

# Polynomial kernel

```python
#Train the model using Support Vector Machine with Polynomial kernel
model_poly = SVC(kernel= "poly",random_state = 0)
model_poly.fit(x_train, y_train)

y_pred_poly = model_poly.predict(x_test)

accuracy_poly = accuracy_score(y_test, y_pred_poly)
precision_poly = precision_score(y_test, y_pred_poly, average='weighted')
recall_poly = recall_score(y_test, y_pred_poly, average='weighted')
f1_poly = f1_score(y_test, y_pred_poly, average='weighted')
```

```python
print("Accuracy for Polynomial Kernel: ", accuracy_poly)
print("Precision for Polynomial Kernel: ", precision_poly)
print("Recall for Polynomial Kernel: ", recall_poly)
print("F1 Score for Polynomial Kernel: ", f1_poly)
```

# Sigmoid Kernel

```python
#Train the model using Support Vector Machine with Sigmoid kernel
model_sigmoid = SVC(kernel="sigmoid", random_state=0)
model_sigmoid.fit(x_train, y_train)

y_pred_sigmoid = model_sigmoid.predict(x_test)

accuracy_sigmoid = accuracy_score(y_test, y_pred_sigmoid)
precision_sigmoid = precision_score(y_test, y_pred_sigmoid, average='weighted')
recall_sigmoid = recall_score(y_test, y_pred_sigmoid, average='weighted')
f1_sigmoid = f1_score(y_test, y_pred_sigmoid, average='weighted')
```

```python
print("Accuracy for Sigmoid Kernel: ", accuracy_sigmoid)
print("Precision for Sigmoid Kernel: ", precision_sigmoid)
print("Recall for Sigmoid Kernel: ", recall_sigmoid)
print("F1 Score for Sigmoid Kernel: ", f1_sigmoid)
```
